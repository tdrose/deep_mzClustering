```python
from mz_clustering.clustering import Clustering
from metaspace import SMInstance
import umap
import numpy as np
import matplotlib.pyplot as plt
import torch
```

```python
sm = SMInstance()
dsid = "2019-08-02_22h43m20s"
ds = sm.dataset(id=dsid)
tmp = ds.all_annotation_images(fdr=0.2, database=("HMDB", "v4"))
tmp = np.array([x._images[0] for x in tmp])
mheight = tmp.shape[1]
mwidth = tmp.shape[2]
```

```python
clustering = Clustering(tmp,
                        num_cluster=7,
                        height=mheight, width=mwidth,
                        lr=0.0001,
                        batch_size=50,
                        knn=True, k=10,
                        use_gpu=False)
```

```python
clustering.image_data.reshape((500, -1)).shape
```

```python
reducer = umap.UMAP()
embedding = reducer.fit_transform(clustering.image_data.reshape((500, -1)))
embedding.shape
```

```python
plt.scatter(
    embedding[:, 0],
    embedding[:, 1])
```

```python
cae, clust = clustering.train()
```

```python
test_x = torch.Tensor(clustering.image_data)
test_x = test_x.reshape((-1, 1, clustering.height, clustering.width))

x = cae.forward(test_x)
pred = clust.forward(test_x)

pred_numpy = pred.cpu().detach().numpy()
```

```python
reducer = umap.UMAP()
embedding_pred = reducer.fit_transform(pred_numpy)
embedding_pred.shape
```

```python
plt.scatter(
    embedding_pred[:, 0],
    embedding_pred[:, 1])
```
